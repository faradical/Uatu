```python
soup = BS(resp.content, "html.parser")
container = soup.find("div", attrs={"id": "placardContainer"})
lis = container.find_all("li", class_="mortar-wrapper")

listings = []

for li in lis:
    # Link
    try:
        link = li.find("a", class_="property-link")["href"]
    except:
        link = ''
    
    # Title
    try:
        title = li.find("div", class_="property-title")['title']
    except:
        title = ''
    
    # Address
    try:
        address = li.find("div", class_="property-address js-url")["title"]
    except:
        address = ''
    
    # Price
    try:
        price = li.find("div", class_="price-range").text
    except:
        price = ''
    
    # Beds
    try:
        beds = li.find("div", class_="bed-range").text
    except:
        beds = ''
    
    # Amenities
    try:
        amenities = li.find("div", class_="property-amenities").get_text().replace("\n",", ")[2:-2]
    except:
        amenities = ''
    
    # Availability
    try:
        availability = li.find("div", class_="availability").text
    except:
        availability = ''
        
    listings.append({
        "link": link,
        "title": title,
        "address": address,
        "price": price,
        "beds": beds,
        "amenities": amenities,
        "availability": availability
    })

pd.DataFrame(listings)
```
